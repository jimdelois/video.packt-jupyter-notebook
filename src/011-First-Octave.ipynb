# First Tests of the Octave Kernal

For starters, let us create an **identity** matrix for the two dimensional space $\mathbb{R}^2$.

````octave
I = [1,0;0,1]
````

And now another identity matrix for the space $\mathbb{R}^3$.

````octave
I3 = [1,0,0;0,1,0;0,0,1]
````

Let's create a $3$ dimensional vector $\vec{v}$ on which this $3 \times 3$ matrix can operate...

````octave
v = [3;4;5]
````

Now let us prove that the $3 \times 3$ identity matrix results in $\vec{v}$ when applied to $\vec{v}$.

````octave
I3 * v
````

So now let us define a 90-degree rotation transform in $\mathbb{R}^2$, counter-clockwise, and then apply it to a simple two-dimensional vector:

````octave
ccw90 = [0,-1;1,0]

ccw90 * [1;5]
````

So now let us note that matrix transformations may be **composed** to result in another singular transformation matrix. Consider this case, in which we compose the `CCW90` operation twice for a full rotation of `180` over the origin:

````octave
ccw180 = ccw90 * ccw90

ccw180 * [1;5]
````

## Unrelated
Apropos of absolutely nothing, here is an interesting built-in graph of the Octave library, known as the "Sombrero" plot.

While the graph below is created simply by invoking the command `sombrero`, the actual underlying equation is something along the lines of:
```
z = sin (sqrt (x^2 + y^2)) / (sqrt (x^2 + y^2))
```
or, in $\LaTeX$:
$$z = f(x,y) = \frac{\sin({\sqrt{x^2 + y^2}})}{\sqrt{x^2 + y^2}}$$

````octave
sombrero
````
